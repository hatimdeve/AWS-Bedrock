```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils import to_categorical
from keras.layers import Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
import cv2
from sklearn.model_selection import train_test_split
import pickle
import os
import pandas as pd
import random
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
def loading_the_data(data_dir):
    filepaths = []
    labels = []

    # Get folder names
    folds = os.listdir(data_dir)

    for fold in folds:
        foldpath = os.path.join(data_dir, fold)
        filelist = os.listdir(foldpath)
        for file in filelist:
            fpath = os.path.join(foldpath, file)

            filepaths.append(fpath)
            labels.append(fold)

    # Concatenate data paths with labels into one DataFrame
    Fseries = pd.Series(filepaths, name='filepaths')
    Lseries = pd.Series(labels, name='labels')

    df = pd.concat([Fseries, Lseries], axis=1)

    return df
```

```python
data_dir = '/content/my_data/myData'
df = loading_the_data(data_dir)
df
```

```python
import cv2
import numpy as np
import pandas as pd

# Chargez votre DataFrame contenant les chemins d'accs aux images et les labels
# Assurez-vous que votre DataFrame a les colonnes 'filepaths' et 'label'
# Remplacez df par le nom de votre DataFrame

# Initialisation des listes
images = []
classNo = []

print("Importing Images and Labels...")

# Parcours de chaque ligne de la DataFrame
for index, row in df.iterrows():
    # Chemin d'accs  l'image
    img_path = row['filepaths']
    # Label de classe
    label = row['labels']

    # Lecture de l'image depuis le chemin d'accs
    curImg = cv2.imread(img_path)
    # Redimensionnement de l'image  la taille spcifie (32x32 pixels)
    curImg = cv2.resize(curImg, (32, 32))

    # Ajout de l'image  la liste des images
    images.append(curImg)
    # Ajout de l'indice de classe correspondant  la liste des classes
    # Vous pouvez utiliser un dictionnaire pour mapper les noms de classe  des indices de classe
    class_mapping={'20': 0,
               '30': 1,
               '40': 2,
               '50': 3,
               '60': 4,
               '70': 5,
               '80': 6,
               '90': 7,
               '100': 8,
               '110': 9,
               '120': 10}

    # Remplacez class_mapping par votre dictionnaire de mappage
    classNo.append(class_mapping[label])

# Conversion des listes en tableaux numpy
images = np.array(images)
classNo = np.array(classNo)
```

```python
import matplotlib.pyplot as plt

# Fonction pour convertir l'image en niveaux de gris
def grayscale(img):
    if len(img.shape) > 2:  # Vrifier si l'image n'est pas dj en niveaux de gris
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

# Fonction pour galiser l'histogramme de l'image (si elle est en niveaux de gris)
def equalize(img):
    if len(img.shape) > 2:  # Vrifier si l'image est en niveaux de gris
        img = cv2.equalizeHist(img)
    return img

# Fonction pour prtraiter l'image
def preprocessing(img):
    img = grayscale(img)     # Convertir en niveaux de gris
    img = equalize(img)      # galiser l'histogramme
    img = img / 255          # Normaliser les valeurs entre 0 et 1
    return img

# Appliquer le prtraitement  chaque ensemble d'images
X_train = np.array([preprocessing(img) for img in X_train])
X_validation = np.array([preprocessing(img) for img in X_validation])
X_test = np.array([preprocessing(img) for img in X_test])

# Vrifier les proprits des images aprs prtraitement
print("Proprits des images aprs prtraitement :")
print("X_train_preprocessed shape:", X_train.shape)
print("X_train_preprocessed dtype:", X_train.dtype)
print("X_train_preprocessed min value:", np.min(X_train))
print("X_train_preprocessed max value:", np.max(X_train))

# Afficher une image alatoire de X_train aprs prtraitement
plt.imshow(X_train[random.randint(0, len(X_train)-1)], cmap='gray')
plt.axis('off')
plt.show()
```

```python
X_train=X_train.reshape(X_train.shape[0],X_train.shape[1],X_train.shape[2],1)
X_validation=X_validation.reshape(X_validation.shape[0],X_validation.shape[1],X_validation.shape[2],1)
X_test=X_test.reshape(X_test.shape[0],X_test.shape[1],X_test.shape[2],1)
```

```python
myList = os.listdir(data_dir)
# Nombre total de classes
noOfClasses = len(myList)
noOfClasses
```

```python
dataGen= ImageDataGenerator(width_shift_range=0.1,   # 0.1 = 10%     IF MORE THAN 1 E.G 10 THEN IT REFFERS TO NO. OF  PIXELS EG 10 PIXELS
                            height_shift_range=0.1,
                            zoom_range=0.2,  # 0.2 MEANS CAN GO FROM 0.8 TO 1.2
                            shear_range=0.1,  # MAGNITUDE OF SHEAR ANGLE
                            rotation_range=10)  # DEGREES
dataGen.fit(X_train)
batches= dataGen.flow(X_train,y_train,batch_size=20)  # REQUESTING DATA GENRATOR TO GENERATE IMAGES  BATCH SIZE = NO. OF IMAGES CREAED EACH TIME ITS CALLED
X_batch,y_batch = next(batches)

# TO SHOW AGMENTED IMAGE SAMPLES
fig,axs=plt.subplots(1,15,figsize=(20,5))
fig.tight_layout()

for i in range(15):
    axs[i].imshow(X_batch[i].reshape(imageDimesions[0],imageDimesions[1]))
    axs[i].axis('off')
plt.show()




y_train = to_categorical(y_train,noOfClasses)
y_validation = to_categorical(y_validation,noOfClasses)
y_test = to_categorical(y_test,noOfClasses)
```

```python
def myModel():
    model = Sequential()
    model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=X_train.shape[1:]))
    model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(rate=0.25))
    model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
    model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(rate=0.25))
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(rate=0.5))
    model.add(Dense(noOfClasses, activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
import tensorflow as tf
tf.keras.utils.set_random_seed(42)
batch_size_val=10  # how many to process together
steps_per_epoch_val=1000
epochs_val=100
model = myModel()
print(model.summary())
history=model.fit(dataGen.flow(X_train,y_train,batch_size=batch_size_val),steps_per_epoch=steps_per_epoch_val,epochs=epochs_val,validation_data=(X_validation,y_validation),shuffle=1)
```
